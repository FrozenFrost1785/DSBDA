```python
# ye jo upar circles dikh rhe h that are outliers .An outlier is a data point that significantly differs from other values in dataset
```

```python
#now to deal with outlier we will use IQR method
```

```python
def RemoveOutlier(df, var):
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    high = Q3 + 1.5 * IQR
    low = Q1 - 1.5 * IQR
    return df[(df[var] >= low) & (df[var] <= high)]
    
dfnew1 = RemoveOutlier(df1, 'Maths_marks')
sns.boxplot(data=dfnew1, x='Maths_marks')
```
