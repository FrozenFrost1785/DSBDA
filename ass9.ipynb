```python
'''
 Data Visualization II 
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution 
of age with respect to each gender along with the information about whether they survived 
or not. (Column names : 'sex' and 'age') 
2. Write observations on the inference from the above statistics
'''
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

```python
titanic.isnull().sum()
```

```python
fig, axes = plt.subplots(1,2)
sns.boxplot( ax=axes[0],data = titanic, y ='age',)
sns.boxplot( ax=axes[1],data = titanic, y ='fare',)
plt.show()
```

```python
fig, axes = plt.subplots(1,3, sharey=True)
sns.boxplot( ax=axes[0],data = titanic, x='sex', y ='age', hue = 'sex',)
sns.boxplot( ax=axes[1],data = titanic, x='pclass', y ='age', hue = 'pclass',)
sns.boxplot(ax=axes[2],data = titanic, x='survived', y ='age', hue = 'survived', )
plt.show()
# sharey=True tells Matplotlib to: Share the same y-axis scale across all 3 plots.
```

```python
fig, axes = plt.subplots(1,3, sharey=True)
sns.boxplot( ax=axes[0],data = titanic, x='sex', y ='fare', hue = 'sex', log_scale = True)
sns.boxplot( ax=axes[1], data = titanic, x='pclass', y ='fare', hue = 'pclass',log_scale = True)
sns.boxplot(ax=axes[2],data = titanic, x='survived', y ='fare', hue = 'survived',  log_scale = True)
plt.show()

#  Why use log_scale=True:
#    The fare column in the Titanic dataset has a few extremely high values (outliers), while most values are much lower. This causes:
#    A compressed boxplot, where you canâ€™t see the details of most data points.
#    By applying a log scale:
#    It spreads out the smaller values.
#    Compresses the large outliers.
#    Makes the plot easier to read and interpret.
```

```python
fig, axes = plt.subplots(1,3, sharey=True)
sns.boxplot( ax=axes[0],data = titanic, x='sex', y ='fare', hue = 'sex')
sns.boxplot( ax=axes[1], data = titanic, x='pclass', y ='fare', hue = 'pclass')
sns.boxplot(ax=axes[2],data = titanic, x='survived', y ='fare', hue = 'survived')
plt.show()
```
