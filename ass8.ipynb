```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

```python
titanic.isnull().sum()
```

```python
fig,axis = plt.subplots(1,3)   #This creates a figure (fig) and a row of 3 subplots (axis is an array of 3 Axes objects).
sns.histplot(ax = axis[0], data = titanic, x='sex', hue = 'sex') #hue='sex'means bars are color based on sex (e.g., male vs. female).
sns.histplot(ax = axis[1], data = titanic, x='sex', )
sns.histplot(ax = axis[2], data = titanic, x='pclass', hue = 'pclass')
plt.show()
```

```python
fig, axis = plt.subplots(1,2)
sns.histplot(ax = axis[0], data = titanic, x='age', kde = True)
sns.histplot(ax = axis[1], data = titanic, x='fare',kde = True)
plt.show()
#Kernel Density Estimate (KDE):
#  In simple terms:
#    A histogram shows bars based on how many values fall into each bin.
#    A KDE curve tries to smooth out those bars into a continuous curve, giving a better sense of the shape of the distribution.
#    KDE works only with numeric/continuous variables, not categorical ones.

```

```python
fig, axis = plt.subplots(2,2)
sns.histplot(ax = axis[0,0], data=titanic, x='age', hue = 'sex', kde = True)
# by above command:
#   Plots a histogram of passenger ages (x='age') on the subplot at axis[0,0].
#   Uses different colors for male and female passengers (hue='sex').
#   Adds a KDE curve for each sex group (kde=True).
sns.histplot(ax = axis[0,1], data=titanic, x='fare', hue = 'sex',kde = True)
sns.histplot(ax = axis[1,0], data=titanic, x='age', hue = 'survived',kde= True)
sns.histplot(ax = axis[1,1], data=titanic, x='fare', hue='survived',kde = True)
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.histplot(ax=axis[0,0], data=titanic, x='sex', hue='survived',kde = True)
sns.histplot(ax=axis[0,1], data=titanic, x='pclass',hue='survived',kde= True)
sns.histplot(ax=axis[1,0], data=titanic, x='age', hue='survived',kde = True)
sns.histplot(ax=axis[1,1], data=titanic, x='fare', hue='survived', kde = True)
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.histplot(ax=axis[0,0], data=titanic, x='sex', hue='survived',kde = True,shrink=0.8)
sns.histplot(ax=axis[0,1], data=titanic, x='pclass',hue='survived',kde= True,shrink=0.8)
sns.histplot(ax=axis[1,0], data=titanic, x='age', hue='survived',kde = True,shrink=0.8)
sns.histplot(ax=axis[1,1], data=titanic, x='fare', hue='survived', kde = True,shrink=0.8)
plt.show()

#What does shrink do?
#    shrink=1.0 â†’ bars touch each other (default).
#    shrink=0.8 â†’ bars are 80% of the bin width, leaving 20% space between them.
#    shrink=0.5 â†’ even more space between bars.
```
