```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\Iris.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, axis = plt.subplots(2,2)
sns.boxplot(ax = axis[0,0], data = df, y='SepalLengthCm')
sns.boxplot(ax = axis[0,1], data = df, y='SepalWidthCm')
sns.boxplot(ax = axis[1,0], data = df, y='PetalLengthCm')
sns.boxplot(ax = axis[1,1], data = df, y='PetalWidthCm')
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.boxplot(ax = axis[0,0], data = df, x='SepalLengthCm')
sns.boxplot(ax = axis[0,1], data = df, x='SepalWidthCm')
sns.boxplot(ax = axis[1,0], data = df, x='PetalLengthCm')
sns.boxplot(ax = axis[1,1], data = df, x='PetalWidthCm')
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.boxplot(ax = axis[0,0], data = df, y='SepalLengthCm', hue='Species')
sns.boxplot(ax = axis[0,1], data = df, y='SepalWidthCm', hue='Species')
sns.boxplot(ax = axis[1,0], data = df, y='PetalLengthCm', hue='Species')
sns.boxplot(ax = axis[1,1], data = df, y='PetalWidthCm', hue='Species')
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.histplot(ax = axis[0,0], data = df, x='SepalLengthCm', shrink = 0.8, kde = True)
sns.histplot(ax = axis[0,1], data = df, x='SepalWidthCm', shrink = 0.8, kde = True)
sns.histplot(ax = axis[1,0], data = df, x='PetalLengthCm', shrink = 0.8, kde = True)
sns.histplot(ax = axis[1,1], data = df, x='PetalWidthCm', shrink = 0.8, kde = True)
plt.show()
```

```python
fig, axis = plt.subplots(2,2)
sns.histplot(ax=axis[0,0], data=df, x='SepalLengthCm', hue='Species', shrink=0.8, kde= True)
sns.histplot(ax=axis[0,1], data = df, x='SepalWidthCm', hue = 'Species', element = 'poly', shrink = 0.8, kde = True)
sns.histplot(ax=axis[1,0], data = df, x='PetalLengthCm', hue = 'Species', element = 'poly', shrink = 0.8, kde = True)
sns.histplot(ax=axis[1,1], data = df, x='PetalWidthCm', hue = 'Species', element = 'poly', shrink = 0.8, kde = True)
plt.show()
# element='poly' means: Draw each histogram as a smooth polygon line instead of rectangular bars.
```
